```python
import pandas as pd

# Load the dataset
df = pd.read_csv('goemotions_2.csv')

# Define the grouping of GoEmotions labels to the 7 target categories
emotion_groups = {
    'Angry': ['anger', 'annoyance', 'disapproval'],
    'Disgust': ['disgust'],
    'Fear': ['fear', 'nervousness'],
    'Happy': ['joy', 'amusement', 'approval', 'excitement', 'gratitude', 'love', 'optimism', 'pride', 'relief', 'admiration', 'caring', 'desire'],
    'Neutral': ['neutral'],
    'Sad': ['sadness', 'disappointment', 'embarrassment', 'grief', 'remorse'],
    'Surprise': ['surprise', 'realization', 'confusion', 'curiosity']
}

# Define the numerical encoding (0-6)
emotion_encoding = {
    'Angry': 0,
    'Disgust': 1,
    'Fear': 2,
    'Happy': 3,
    'Neutral': 4,
    'Sad': 5,
    'Surprise': 6
}

# Process the data to extract the 7 emotions
rows = []
for idx, row in df.iterrows():
    text = row['text']
    # Check each group
    assigned = False
    for emotion, labels in emotion_groups.items():
        # Check if any label in this group is 1
        if any(row[label] == 1 for label in labels if label in row):
            rows.append({
                'TEXT': text,
                'EMOTION': emotion,
                'Emotion_Encoded': emotion_encoding[emotion]
            })
            assigned = True
            break # Take the first matching emotion group

# Create the new DataFrame
new_df = pd.DataFrame(rows)

# Save to CSV (as per system guidelines for user download)
new_df.to_csv('goemotions_final_7.csv', index=False)

# Also save to Excel as requested by the user
try:
    new_df.to_excel('goemotions_final_7.xlsx', index=False)
    excel_success = True
except Exception as e:
    excel_success = False
    error_msg = str(e)

print(f"Total rows processed: {len(new_df)}")
print(new_df.head())
print(f"Excel save status: {excel_success}")
if not excel_success:
    print(f"Excel Error: {error_msg}")
```
